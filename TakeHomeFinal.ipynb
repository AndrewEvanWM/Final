```python
#a

plt.clf()


#function
def function(x1, x2):
    #function is simply if x1+x2 > 25,
    #return 1, otherwise return -1
    y = x1 + x2
    if y > 25:
        return 1
    return -1

#creating the dataset
n = 20
m = 3
d = [None]*n
for i in range(0,20):
    x = [None]*m
    for j in range(0,m-1):
        x[j] = random.randint(0,25)
    x[2] = function(x[0],x[1])
    if (x[2] == 1):
        plt.scatter(x[0],x[1], c='b', label = x[2])
    else:
        plt.scatter(x[0],x[1], c='r', label = x[2])
    d[i] = x

    
#print(d)

plt.title('Linearly Seperable Data')
plt.xlabel("X1")
plt.ylabel("X2")

x1 = numpy.array(range(25))
y1 = 25 - x1
plt.plot(x1, y1, linestyle='solid', color='black')
plt.show()
```
